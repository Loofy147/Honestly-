# Suffix Smoother v0.4.0: Industrial Scale Demo

This notebook demonstrates the industrial-grade features of Suffix Smoother v0.4.0:
1. **Post-Training Pruning** (90% reduction at scale).
2. **Auto-Calibration** via Temperature Scaling.
3. **JSON Portability** for production deployment.

```python
!pip install suffix-smoother==0.4.0
import numpy as np
import time
from suffix_smoother import SuffixSmoother, SuffixConfig

print("✓ Suffix Smoother v0.4.0 loaded.")
```

## 1. Scale Benchmarking (100,000 samples)

```python
rng = np.random.default_rng(42)
data = [(tuple(rng.integers(0, 5, size=5).tolist()), int(rng.integers(0, 10))) for _ in range(100000)]
val_data = data[80000:]
train_data = data[:80000]

model = SuffixSmoother(SuffixConfig(n_classes=10))
t0 = time.time()
model.train(train_data)
print(f"Trained on 80k samples in {time.time()-t0:.2f}s")
print(f"Nodes: {model.n_nodes}")
```

## 2. Calibration & Pruning

```python
print("Optimal T:", model.fit_temperature(val_data))
res = model.prune(min_count=5)
print(f"Pruned {res['removed']} rare nodes.")
print(f"Remaining nodes: {model.n_nodes}")
```

## 3. Production JSON Persistence

```python
model.to_json("model.json")
print("Model saved to JSON. Safe for cross-version deployment.")
```
